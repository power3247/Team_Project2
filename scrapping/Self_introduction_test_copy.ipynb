```python
contents = []

for scraping in introduction_page:
    company = introduction_page[0].select('div.viewTitWrap')[0].text.strip()
    score = introduction_page[0].select('div.adviceTotal > div > span')[0].text.strip()
    group = introduction_page[0].select('div.adviceTotal > p')[0].text.strip()
    Q1 = introduction_page[0].select('dt:nth-child(1) > button > span.tx')[0].text.strip()
    A1 = introduction_page[0].select('dd:nth-child(2) > div.tx')[0].text.strip()
    Q2 = introduction_page[0].select('dt:nth-child(3) > button > span.tx')[0].text.strip()
    A2 = introduction_page[0].select('dd:nth-child(4) > div.tx')[0].text.strip()
    Q3 = introduction_page[0].select('dt:nth-child(5) > button > span.tx')[0].text.strip()
    A3 = introduction_page[0].select('dd:nth-child(6) > div.tx')[0].text.strip()
    Q4 = introduction_page[0].select('dt:nth-child(7) > button > span.tx')[0].text.strip()
    A4 = introduction_page[0].select('dd:nth-child(8) > div.tx')[0].text.strip()
    result = [company, score, group, Q1, A1, Q2, A2, Q3, A3, Q4, A4 ]
    contents.append(result)

contents
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(contents,columns=['company', 'score', 'group', 'Q1', 'A1', 'Q2', 'A2', 'Q3', 'A3', 'Q4', 'A4'])
```
